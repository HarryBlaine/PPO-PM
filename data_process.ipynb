```python
import datetime as dt
import pandas as pd
import numpy as np
```

```python
Stock_Price = pd.read_csv('all_stocks_5yr.csv')
Stock_Price
```

```python
count = Stock_Price["Name"].value_counts()
dataframe = count.to_frame()
dataframe
```

```python
dataframe = dataframe[(dataframe['Name'] != 1259)]
dataframe["name"] = dataframe.index
dataframe
```

```python
name=dataframe["name"].values
len(name)
```

```python
for i in name:
    Stock_Price = Stock_Price.drop(Stock_Price[Stock_Price["Name"] == i].index)
Stock_Price
```
