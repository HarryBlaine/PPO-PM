```python
import datetime as dt
import pandas as pd
import numpy as np
```

```python
Stock_Price = pd.read_excel('./Stock Market/FTSE100.xlsx')
Stock_Price
```

```python
Stock_Price = Stock_Price.drop(["amt"], axis=1)
Stock_Price
```

```python
for i in range(100):
    if i>0:
        print(i)
        Stock_Price = Stock_Price.drop(["amt."+str(i)], axis=1)
```

```python
Stock_Price
```

```python
Stock_Price = Stock_Price.dropna(axis=1,how="all")
Stock_Price
```

```python
Stock_Price.to_csv("FTSE_first.csv",index=False,header=True)
```
