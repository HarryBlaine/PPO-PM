```python
import datetime as dt
import pandas as pd
import numpy as np
```

```python
open = pd.read_excel('./Stock Market/NIKKEI open.xlsx')
open
```

```python
name = open.columns
name
```

```python
for n in name:
    if n != "Date":
        Stock_Price[n + "-volume"] = Stock_Price[n + "-volume"]/100000
```

```python
Stock_Price.to_csv("./Stock Market/HangSeng_final.csv",index=False,header=True)
```
